```python
# Get total number of elements
total_elements = train_atac.X.shape[0] * train_atac.X.shape[1]

# Get number of non-zero (stored) elements
non_zeros = train_atac.X.nnz

# 1. Sparsity (percentage of ZEROS)
sparsity = 1 - (non_zeros / total_elements)
print(f"Sparsity: {sparsity:.4f} (or {sparsity * 100:.2f}%)")

# 2. Density (percentage of NON-ZEROS)
density = non_zeros / total_elements
print(f"Density: {density:.4f} (or {density * 100:.2f}%)")
```
